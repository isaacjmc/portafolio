```python
import git
from datetime import datetime
import os
# Configuración inicial, reemplaza con tus valores de acuerdo a tu repositorio local
#Importante haber configurado previamente el repositorio local con el repositorio remoto, para darle la llave a hacer cambios automaticos desde el dispositivo que lo ejecute
RUTA_REPOSITORIO = r"C:\Users\User\Documents\GitHub\Subidas"
REMOTO_NOMBRE = 'origin'
RAMA_NOMBRE = 'main' 
MENSAJE_COMMIT = f"Commit automático desde Python - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"


def git_automatico(ruta_repo, mensaje_commit, remoto_nombre, rama_nombre):
    try:
        repo = git.Repo(ruta_repo)
        remote = repo.remote(name=remoto_nombre)
        
        print(f"Repositorio cargado: {os.path.basename(ruta_repo)}")
        print("-" * 30)

        print("Preparando 'git add' y 'git commit'...")
        
        repo.git.add(A=True)
        
        # Verificar si hay cambios en el index para commitear
        if repo.index.diff("HEAD"):
            print("Archivos detectados en el área de preparación (staging).")
            
            # Crear el commit
            commit = repo.index.commit(mensaje_commit)
            print(f"Commit exitoso: {commit.hexsha[:7]}")
            
            # --- 2. PUSH ---
            print(f"Ejecutando 'git push {remoto_nombre} {rama_nombre}'...")
            
            # Subir los commits al remoto
            remote.push(refspec=rama_nombre)
            print("Push completado. Cambios subidos a GitHub.")
        else:
            print("No hay cambios para commitear. Saltando commit y push.")
                
    except git.GitCommandError as git_err:
        print(f"Error CRÍTICO de Git (Comando): {git_err}")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")

# Ejecutar la función
git_automatico(RUTA_REPOSITORIO, MENSAJE_COMMIT, REMOTO_NOMBRE, RAMA_NOMBRE)
```
